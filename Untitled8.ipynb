```python
import pandas as pd
import matplotlib.pyplot as plt

Data = {
    "Date": ["2025-01-05", "2025-01-12", "2025-02-08","2025-02-15", "2025-03-10", "2025-03-20", "2025-04-05", "2025-04-15" ],

    "Product" : ["Laptop", "Mouse", "Laptop", "Keyboard", "Mouse", "Laptop", "Keyboard", "Mouse"],

    "Quantity" : [2,5,1,3,4,2,2,6],

    "Price": [50000,500,50000,1500,500,50000,1500,500]}

df = pd.DataFrame(Data)
print(df)

df.drop_duplicates(inplace=True)
df["Date"] = pd.to_datetime(df["Date"])


# To find total sales
df["Total_Sales"] = df["Quantity"] * df["Price"]

# To find top Products
top_products = df.groupby("Product")["Total_Sales"].sum()
top_products = top_products.sort_values(ascending=False)

print(top_products)


# To find Monthly sales
df["Month"] = df["Date"].dt.month_name()
monthly_sales = df.groupby("Month") ["Total_Sales"].sum()
print(monthly_sales)


# Bar Chart
plt.figure(figsize=(6,4))

top_products.plot(kind="bar")
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title("Top Products")
plt.show()

#Line chart
plt.figure(figsize=(6,4))

monthly_sales.plot(kind="line", marker="o")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.show()







```
